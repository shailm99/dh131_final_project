```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/SIMPLIFIED_with_locations.csv', index_col = 0)
```

```python
df2 = df['Captives arrived at 1st port'].isnull().groupby([df.beg_loc]).sum().astype(int)
#sns.displot(df2['count'])
df2 = pd.DataFrame(df2.sort_values(ascending=False))
```

```python
df2['total_count'] = df.beg_loc.value_counts()
```

```python
df2['percentage_null'] = df2['Captives arrived at 1st port'] / df2['total_count']
```

```python
df2[df2.total_count > 10].sort_values(by = 'percentage_null', ascending = False)
```

```python
plt.figure(figsize = (16,8))
sns.distplot(df2.percentage_null)
plt.xlim((0,1))
plt.xlabel("Percentage of Rows without Number of Captives Reported by Location")
plt.title("There appears t")
```
