```python
# we will drop the rows that are missing location data 

ships.dropna(subset=['Voyage itinerary imputed port where began (ptdepimp) place',
       'Voyage itinerary imputed principal place of slave purchase (mjbyptimp) ',
       'Voyage itinerary imputed principal port of slave disembarkation (mjslptimp) place'], inplace=True)
```

```python
ships['Voyage itinerary imputed port where began (ptdepimp) place'].apply(lambda x: str(x).split(',')[0])
```

```python
def clean_locations(df):
    df['Voyage itinerary imputed port where began (ptdepimp) place'] = df['Voyage itinerary imputed port where began (ptdepimp) place'].apply(lambda x: str(x).split(',')[0])
    df['Voyage itinerary imputed principal place of slave purchase (mjbyptimp) '] = df['Voyage itinerary imputed principal place of slave purchase (mjbyptimp) '].apply(lambda x: str(x).split(',')[0])
    df['Voyage itinerary imputed principal port of slave disembarkation (mjslptimp) place'] = df['Voyage itinerary imputed principal port of slave disembarkation (mjslptimp) place'].apply(lambda x: str(x).split(',')[0])
    return df
```

```python
ships = clean_locations(ships)
```

```python
ships
```

```python
rename_columns = {'Voyage itinerary imputed port where began (ptdepimp) place' : 'beg_loc',
       'Voyage itinerary imputed principal place of slave purchase (mjbyptimp) ' : 'purchase_loc',
       'Voyage itinerary imputed principal port of slave disembarkation (mjslptimp) place' : 'end_loc'}
ships.rename(columns = rename_columns, inplace = True)
```

```python
ships
```

```python
def clean_locationsv2(df):
    df['beg_loc'] = df['beg_loc'].apply(lambda x: str(x).replace(', port unspecified', ''))
    df['purchase_loc'] = df['purchase_loc'].apply(lambda x: str(x).replace(', port unspecified', ''))
    df['end_loc'] = df['end_loc'].apply(lambda x: str(x).replace(', port unspecified', ''))
    return df
```

```python
ships = clean_locationsv2(df=ships)
```

```python
from geopy.geocoders import Nominatim

# Initialize Nominatim API
geolocator = Nominatim(user_agent="MyApp")

location = geolocator.geocode("Hyderabad")
```

```python
unique_locations = set(ships['beg_loc'].unique().tolist() + ships['purchase_loc'].unique().tolist() + ships['end_loc'].unique().tolist())
```

```python
mapping = {}
no_location = []
for loc in unique_locations:
    try:
        location = geolocator.geocode(loc)
        mapping[loc] = (location.latitude, location.longitude)
    except:
        no_location.append(loc)
```

```python
len(mapping)
```

```python
len(unique_locations)
```

```python
# there are 60 locations that cannot be found so we can just drop those rows 
```

```python
def extract_long_lats(df, col):
    lat_long = ships[col].apply(lambda x: mapping.get(x, (None, None)))
    df[f'{col}_lat'] = lat_long.apply(lambda x: x[0])
    df[f'{col}_long'] = lat_long.apply(lambda x: x[1])
    return df
```

```python
ships = extract_long_lats(ships, 'beg_loc')
ships = extract_long_lats(ships, 'purchase_loc')
ships = extract_long_lats(ships, 'end_loc')
ships
```
